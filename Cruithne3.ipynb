```python
#COROTATIONAL
num_years = 660
num_steps = num_years*365 # each step is a day for this simulation

x_cha = 0
y_cha = 0


for i in range(num_steps):
#    initial_x = sim.particles[3].x
#    initial_y = sim.particles[3].y
    
    sim.dt = sim.particles[3].P/365. # set time_step size to be a single earth day
    sim.step() # advance the simulation by a single step
    
#    new_x = sim.particles[3].x
#    new_y = sim.particles[3].y
    
#    x_cha = initial_x - new_x
#    y_cha = initial_y - new_y
    
#    if new_x < initial_x:
#        x_cha = initial_x - new_x
#    elif new_x > initial_x:
#        x_cha = -1*(initial_x - new_x)
    
    cruithne_values = sim.particles[-1] 
    earth_values = sim.particles[3] 
    sun_values = sim.particles[0] 
    
    # always adding x value to get back to 1
#    if earth_values.x < 0:
#        x_disp = 1 + abs(earth_values.x) # x displacement of earth from ideal
#    elif earth_values.x >= 0:
#        x_disp = 1 - earth_values.x # x displacement of earth from ideal
    # always adding y value to get back to 0
#    y_disp = -1 * (earth_values.y - 0) # y displacement of earth from ideal
    
    # since each step is a single day, we can take it to be 2pi/365 per day therefore theta = (2pi/365)*i    
    theta = (2*(math.pi)/365)*i
    # subtraction of math.pi fractions was used to rotate the schema approximately for prettier presentation
    x_data1.append((cruithne_values.x*math.cos(-theta)) - (cruithne_values.y*math.sin(-theta))) # cruithne in rotational frame
    y_data1.append((cruithne_values.x*math.sin(-theta)) + (cruithne_values.y*math.cos(-theta)))
    x_data2.append((earth_values.x*math.cos(-theta)) - (earth_values.y*math.sin(-theta))) # earth in rotational frame
    y_data2.append((earth_values.x*math.sin(-theta)) + (earth_values.y*math.cos(-theta)))
    x_data3.append((sun_values.x*math.cos(-theta)) - (sun_values.y*math.sin(-theta))) # sun in rotational frame -- still at origin
    y_data3.append((sun_values.x*math.sin(-theta)) + (sun_values.y*math.cos(-theta)))
    
#    x_hold = earth_values.x
#    y_hold = earth_values.y

plotted_steps += num_steps
```
