```python
times = 4*365.
num_steps = np.linspace(0,times,1000)

#sim.particles[-1].m = 0.4

for step in num_steps:
    sim.dt = sim.particles[3].P/365.
    sim.integrate(step)
#    sim.step()
    x_1.append(sim.particles[1].x) #mercury
    y_1.append(sim.particles[1].y) 
    x_2.append(sim.particles[2].x) #venus
    y_2.append(sim.particles[2].y)
    x_3.append(sim.particles[3].x) #earth
    y_3.append(sim.particles[3].y)
    x_4.append(sim.particles[4].x) #mars
    y_4.append(sim.particles[4].y)
    x_5.append(sim.particles[5].x) #Jupiter
    y_5.append(sim.particles[5].y)
    x_6.append(sim.particles[-1].x) #satellite
    y_6.append(sim.particles[-1].y)


#plotted_steps += num_steps
```
