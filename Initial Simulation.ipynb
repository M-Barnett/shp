```python
import rebound
import numpy as np
%matplotlib inline
import matplotlib.pylab as plt
```

```python
def setup():
    sim = rebound.Simulation()
    sim.add(m=1)
    sim.add(m=1e-3,a=1)
    sim.add(m=1e-3,a=2)
    return sim
```

```python
times = np.linspace(0., 100, 500)
eccs = np.zeros((len(times), 2))
```

```python
sim = setup()
for i, t in enumerate(times):
    sim.integrate(t)
    eccs[i,0] = sim.particles[1].e
    eccs[i,1] = sim.particles[2].e
```

```python
plt.plot(times,eccs)
```
