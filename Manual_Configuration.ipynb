```python
sim=rebound.Simulation()

xdct = []
ydct = []
#7,12,28

for i in range(1):
    for j in range(12):
        for k in range(2):
            chosen_date = "" + str(2001+i+1) + "-" + str(j+1) + "-" + str((k+1)*8) + " 13:00"
            sim.add("NAME= 1986 TO;",date=chosen_date)
            xdct.append(sim.particles[-1].x)
            ydct.append(sim.particles[-1].y)
            sim.remove(index=0)

print(xdct)
print(ydct)
```

```python
sim=rebound.Simulation()

xdet = []
ydet = []
#7,12,28

for i in range(1):
    for j in range(12):
        for k in range(2):
            chosen_date = "" + str(2001+i+1) + "-" + str(j+1) + "-" + str((k+1)*8) + " 13:00"
            sim.add("Earth",date=chosen_date)
            xdet.append(sim.particles[-1].x)
            ydet.append(sim.particles[-1].y)
            sim.remove(index=0)

print(xdet)
print(ydet)
```

```python
fig = plt.figure(figsize=(18, 16))
ax = fig.add_subplot(1,1,1)
ax.plot(xdct, ydct, color='tab:blue', marker='o')
ax.plot(xdet, ydet, color='tab:green', marker='o')
#fig, ax = plt.subplots()
#ax.plot(x_list, y_list)
#ax.set_title("Corotational " + str(plotted_steps) + " days")
ax.set_xlim([-3.,3.])
ax.set_ylim([-3.,3.])
plt.show()
```

```python
xcomb = [0]*len(xdct)
ycomb = [0]*len(xdct)
for i in range(len(xdct)):
    xcomb[i] = xdct[i] - xdet[i]
    ycomb[i] = ydct[i] - ydet[i]
```

```python
fig = plt.figure(figsize=(18, 16))
ax = fig.add_subplot(1,1,1)
ax.plot(xcomb, ycomb, color='tab:blue', marker='o')

#fig, ax = plt.subplots()
#ax.plot(x_list, y_list)
#ax.set_title("Corotational " + str(plotted_steps) + " days")
ax.set_xlim([-3.,3.])
ax.set_ylim([-3.,3.])
plt.show()
```
